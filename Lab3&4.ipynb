#Bi-gram

```python
import nltk
from nltk import bigrams

bi_grams = list(bigrams(words))

bi_gram_freq = Counter(bi_grams)

top_10_bi_grams = bi_gram_freq.most_common(10)

for bi_gram, freq in top_10_bi_grams:
    print(f"{bi_gram}: {freq} lần")
```

```python
conditional_probs = {}

for (w1, w2), count in bi_gram_freq.items():
    conditional_prob = count / word_freq[w1]
    conditional_probs[(w2, w1)] = conditional_prob

for (w2, w1), prob in conditional_probs.items():
    print(f"P({w2}|{w1}) = {prob}")
```

```python
print(conditional_probs)
```

```python
def colected_conditional_probs(input_word, conditional_probs):

    colected_pairs = [(key, value) for key, value in conditional_probs.items() if key[1] == input_word]
    return colected_pairs
```

```python
def colected_dict(vocab, conditional_probs):
    colected_dict = {}

    for word in vocab:
        colected_pairs = colected_conditional_probs(word, conditional_probs)
        colected_dict[word] = colected_pairs

    return colected_dict

colected_dict = colected_dict(vocab, conditional_probs)
```

```python
print(colected_dict["lòng"])
```

```python
import random

def generate_next_word(current_word):
    random_value = random.uniform(0, 1)

    cumulative_prob = 0.0
    next_word = None

    for (word, prev_word), prob in colected_dict[current_word]:
        cumulative_prob += prob
        if random_value < cumulative_prob:
            next_word = word
            break

    return next_word
```

#Tri-gram

```python
from nltk import trigrams

tri_grams = list(trigrams(words))

tri_gram_freq = Counter(tri_grams)

top_10_tri_grams = tri_gram_freq.most_common(10)

for tri_gram, freq in top_10_tri_grams:
    print(f"{tri_gram}: {freq} lần")
```

```python
conditional_probs_tri = {}

for (w1, w2, w3), count in tri_gram_freq.items():
    conditional_prob_tri = count / bi_gram_freq[w1,w2]
    conditional_probs_tri[(w3, w1, w2)] = conditional_prob_tri

for (w3, w1, w2), prob in conditional_probs_tri.items():
    print(f"P({w3}|{w1},{w2}) = {prob}")
```

```python
def colected_conditional_probs_tri(input_word_1, input_word_2, conditional_probs_tri):

    colected_pairs = [(key, value) for key, value in conditional_probs_tri.items() if (key[1] == input_word_1 and key[2] == input_word_2)]
    return colected_pairs
```

```python
def colected_dict_tri(vocab_bi, conditional_probs_tri):
    colected_dict_tri = {}

    for (word_1, word_2) in vocab_bi:
        colected_pairs = colected_conditional_probs_tri(word_1, word_2, conditional_probs_tri)
        colected_dict[word_1, word_2] = colected_pairs

    return colected_dict

colected_dict_tri = colected_dict_tri(vocab_bi, conditional_probs_tri)
```

```python
def generate_next_word_tri(current_word_1, current_word_2):
    random_value = random.uniform(0, 1)

    cumulative_prob = 0.0
    next_word = None

    for (word, prev_word_1, prev_word_2), prob in colected_dict_tri[current_word_1, current_word_2]:
        cumulative_prob += prob
        if random_value < cumulative_prob:
            next_word = word
            break

    return next_word
```
