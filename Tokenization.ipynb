# Shakespeare

```python
import re

cleaned_shakes = re.sub(r'[^a-zA-Z\s]', '', shakes)
lower_shakes = cleaned_shakes.lower()
```

```python
tokens_shakes = re.findall(r'\b\w+\b', lower_shakes)
```

```python
print(tokens_shakes)
```

```python
len(tokens_shakes)
```

```python
from collections import Counter
count_shakes = Counter(tokens_shakes)
vocab_shakes = count_shakes.keys()
len(vocab_shakes)
```

```python
print(count_shakes.most_common(10))
```

```python
with open('output_shakes.txt', 'w') as f:
  for word, freq in count_shakes.items():
    f.write(f"{word} , {freq}\n")
```

```python
from nltk import bigrams, trigrams
```

```python
bi_shakes = list(bigrams(tokens_shakes))
count_bi_shakes = Counter(bi_shakes)
print(count_bi_shakes.most_common(10))
```

```python
with open('output_shakes_bi.txt', 'w') as f:
  for word, freq in count_bi_shakes.items():
    f.write(f"{word} , {freq}\n")
```

```python
tri_shakes = list(trigrams(tokens_shakes))
count_tri_shakes = Counter(tri_shakes)
print(count_tri_shakes.most_common(10))
```

```python
with open('output_shakes_tri.txt', 'w') as f:
  for word, freq in count_tri_shakes.items():
    f.write(f"{word} , {freq}\n")
```

```python
import vietokenizer

tokenizer = vietokenizer.vntokenizer()
```

```python
cleaned_kieu = re.sub(r"[0-9\.\,\!\&\#\$\*\n\?\:]+",r" ",truyenkieu)
```

```python
cleaned_kieu[:100]
```

```python
lower_kieu = cleaned_kieu.lower()
```

```python
tokenized_data = tokenizer(lower_kieu)
```

```python
len(tokenized_data)
```

```python
tokens_kieu = tokenized_data.split()
```

```python
len(tokens_kieu)
```

```python
from collections import Counter
```

```python
token_count = Counter(tokens_kieu)
vocab_kieu = token_count.keys()
len(vocab_kieu)
```

```python
print(token_count.most_common(10))
```

```python
with open('output_kieu.txt', 'w') as f:
  for word, freq in token_count.items():
    f.write(f"{word} , {freq}\n")
```

```python
bi_kieu = list(bigrams(tokens_kieu))
count_bi_kieu = Counter(bi_kieu)
print(count_bi_kieu.most_common(10))
```

```python
with open('output_kieu_bi.txt', 'w') as f:
  for word, freq in count_bi_kieu.items():
    f.write(f"{word} , {freq}\n")
```

```python
tri_kieu = list(trigrams(tokens_kieu))
count_tri_kieu = Counter(tri_kieu)
print(count_tri_kieu.most_common(10))
```

```python
with open('output_kieu_tri.txt', 'w') as f:
  for word, freq in count_tri_kieu.items():
    f.write(f"{word} , {freq}\n")
```
